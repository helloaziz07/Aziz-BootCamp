```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Housing.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
#  LINE PLOT

plt.plot(data['area'], data['price'])
plt.xlabel('Area')
plt.ylabel('Price')

plt.title('Line Plot')
plt.show()
```

```python
#  LINE PLOT

plt.plot(data['bedrooms'], data['price'])
plt.xlabel('Bedrooms')
plt.ylabel('Price')

plt.title('Line Plot')
plt.show()
```

```python
#  LINE PLOT

plt.plot(data['bathrooms'], data['price'])
plt.xlabel('bathrooms')
plt.ylabel('price')

plt.title('Line Plot')
plt.show()
```

```python
# BAR PLOT

plt.bar(data['area'], data['price'])
plt.xlabel('Area')
plt.ylabel('Price')

plt.title('Bar Plot')
plt.show()
```

```python
# BAR PLOT

plt.bar(data['bedrooms'], data['price'], label='Price', color='Green')
plt.legend()
plt.xlabel('Bedrooms')
plt.ylabel('Price')

plt.title('Bar Plot')
plt.show()
```

```python
# BAR PLOT

plt.bar(data['bathrooms'], data['price'], label='Price', color='Red')
plt.legend()
plt.xlabel('Bathrooms')
plt.ylabel('Price')

plt.title('Bar Plot')
plt.show()
```

```python
# BAR PLOT

sns.barplot(x='furnishingstatus', y='price', data=data, hue='furnishingstatus')
```

```python
# BAR PLOT

sns.barplot(y='bathrooms', data=data, hue='bathrooms')
```

```python
# SCATTER PLOT

plt.scatter(data["bedrooms"],data["bathrooms"], color='Orange')
plt.xlabel("Bedrooms")
plt.ylabel("Bathrooms")

plt.title('Scatter Plot')
plt.show()
```

```python
# COUNT PLOT

sns.countplot(x='bathrooms', data=data, hue='bedrooms', palette='deep')
```

```python
# BOX PLOT

sns.boxplot(x="price",data=data)
```

```python
# HIST PLOT

sns.histplot(x="bedrooms",data=data, hue="bedrooms",palette="deep")
```
